```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import os
```

```python
data_file_path = os.path.abspath(os.path.join(os.pardir,'data','cleaned_AG.csv'))
export_file_path = os.path.abspath(os.path.join(os.pardir,'data','bag_of_words_training_data.csv'))

# data_file_path = os.path.abspath(os.path.join(os.pardir,'data','cleaned_bing.csv'))
# export_file_path = os.path.abspath(os.path.join(os.pardir,'data','bag_of_words_bing_data.csv'))
```

```python
data = pd.read_csv(data_file_path)
data.head()
```

# Bag of Words

```python
#bag of words
vectorizer = CountVectorizer(max_features=1000)
X = vectorizer.fit_transform(data['Documents'].values)

denselist = X.todense().tolist()
df = pd.DataFrame(denselist, columns=vectorizer.get_feature_names())
bag_of_words = pd.concat([data['News Category'],df],axis=1)
bag_of_words.head()
```

```python
#export bag of words feature data
bag_of_words.to_csv(export_file_path,index=False)
```
